### Introduction
The Horse Colic Dataset is a collection of medical records for horses suffering from colic, a common equine emergency. The dataset aims to predict the outcome (survival or non-survival) of horses based on their medical conditions, treatment, and other factors. With 28 attributes and 368 instances, this dataset provides a comprehensive understanding of equine colic and its management.

### Dataset Information
The Horse Colic Dataset contains 368 instances, each representing a horse's medical record. The dataset includes 28 attributes, which can be categorized into:
- Demographic information (age, hospital number)
- Clinical measurements (rectal temperature, pulse, respiratory rate, etc.)
- Subjective assessments (pain level, mucous membranes, capillary refill time, etc.)
- Treatment and diagnostic information (surgery, nasogastric tube, abdominocentesis, etc.)
- Outcome (lived, died, euthanized)

The dataset also includes information on the type of lesion, its location, and whether surgery was performed. Missing values (NA's) are present in the data, requiring appropriate handling for analysis.

**Was surgery able to save these horses lives who had distended small and large intenstine?**

```python
condition1 =  df['abdomen'].isin(['distend_large','distend_small'])
condition2 = df['surgery']=='yes'

temp1 = df[(condition1) & (condition2)]
pivot = temp1.pivot_table(index='outcome', columns='abdomen', aggfunc='size', fill_value=0)\
                    .reset_index().rename_axis(columns=None)
pivot['Total'] = pivot.set_index('outcome').sum(axis=1).values.tolist()
pivot['distend_large'] = round(pivot['distend_large']/pivot['Total']*100,2)
pivot['distend_small'] = round(pivot['distend_small']/pivot['Total']*100,2)
pivot
```

**INSIGHT**

- If the horse has had distended small intestine and went through a surgery, in 61% of the cases he died and in 49% cases he was euthanized.
- There is also a 61% chance of horse living if he had a distended small intestine and went through with the surgery.

**Is surgery a sure shot way of saving the horse or not?**

```python
pivot = df.pivot_table(index='surgery', columns='outcome', aggfunc='size', fill_value=0).reset_index().rename_axis(columns=None)
pivot['Total'] =  pivot.set_index('surgery').sum(axis=1).values.tolist()
pivot['died'] = round(pivot['died'] / pivot['Total'] * 100,2).values.tolist()
pivot['euthanized'] = round(pivot['euthanized'] / pivot['Total'] * 100,2).values.tolist()
pivot['lived'] = round(pivot['lived'] / pivot['Total'] * 100,2).values.tolist()
pivot['dead'] = pivot['died'] + pivot['euthanized'].values.tolist()
pivot = pivot[['surgery','dead', 'lived', 'Total']]
pivot
```

```python
pivot.set_index('surgery')[['dead', 'lived']].plot(kind='bar')
plt.title('Surgery and Outcome Chart')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

**INSIGHT**

-  57% of the horses who went through the surgery ended up not living any more meaning 19 out of 33 horses.
- 56% of the horses who were not put under the needle lived meaning 14 out of 25 horses.
- Surgery cannot confirm if the horse will live.

**Does abdominal_distention have any impact on the outcome?**

```python

query = '''
with cte as (
select abdominal_distention,  sum(case when outcome = 'died' then 1 else 0 end) as died,
                
                sum(case when outcome = 'euthanized' then 1 else 0 end) as euthanized,
                sum(case when outcome = 'lived' then 1 else 0 end) as lived,
                count(*) as total
from data 
where abdominal_distention is not null 
group by 1)

select abdominal_distention, (round(died*1.0/total*100,2) + 
                            round(euthanized*1.0/total*100,2)) as Died_pct, 
                            round(lived*1.0/total*100,2) as Lived_pct, total from cte
                            order by 3 desc
'''
pd.read_sql(query, conn)
```

**INSIGHT**

- With none or slight abdominal distention, the horse likely to live around 63% of the times.
- With moderate or severe abdominal distention, the horse is most likely to die 72% of the times.

## Data Preprocessing

**Fill Out the Null values**

```python
null = pd.DataFrame(round(df.isnull().sum()/len(df)*100,2)).reset_index()\
            .rename(columns={'index': 'Column', 0:'Null_Pct'}).sort_values(by='Null_Pct', ascending=False)
null.query('Null_Pct!=0')
```

```python
null_cols = df[null.query('Null_Pct!=0')['Column']]
for col in null_cols:
    df[col].fillna('Unknown', inplace=True)
```

```python
null_count = sum(list(df.isnull().sum().values.tolist()))
print(f"There are now {null_count} null values in the dataset after imputing.")
```

**Label Encoding**

```python
label_encoders = {}
mappings = {}

# Apply label encoding and store mappings
for column in cat_cols:
    if column in df.columns:
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le
        # Create the mapping like you requested
        mappings[column] = dict(zip(le.transform(le.classes_), le.classes_))
print('Label Encoding Done and Mappings Saved for later use...')
```

**Scaling**

```python
num_cols = [i for i in num_cols if i not in ('id', 'hospital_number')]
df[num_cols]
```

```python
result_cols = []

for column in num_cols:
    result = anderson(df[column].dropna())
    # Check against critical values
    if result.statistic > result.critical_values[2]:  # Compare with 5% significance level
        result_cols.append(column)   
    else:
        print(f'{column} is normally distributed.')

print("Below are the cols that are not normally distributed")
result_cols
```
